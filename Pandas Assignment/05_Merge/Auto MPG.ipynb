### MPG Cars

```python
#Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
#Import and assign each to a to a variable called cars1 and cars2
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")

print(cars1.head())
print(cars2.head())
```

```python
#fix cars1
cars1 = cars1.loc[:, "mpg":"car"]
cars1.head()
```

```python
#What is the number of observations in each dataset?
print(cars1.shape)
print(cars2.shape)
```

```python
#Join cars1 and cars2 into a single DataFrame called cars
cars = cars1.merge(cars2, on='origin')
cars
```

```python
#Create a random number Series from 15,000 to 73,000 for missing columns.
nr_owners = np.random.randint(15000, high=73001, size=398, dtype='l')
nr_owners
```

```python
#Add the column owners to cars
cars['owners'] = nr_owners
cars.tail()
```
